```r
n = 1000
x = cbind(rep(1, n), rnorm(n, 5, 2))
# x = as.matrix(rnorm(n, 4, 3), n, 1)
y = as.matrix(3*x[,2] + 2, n, 1)

# define loss function ell(beta) = MSE(beta)
# matrix product: %*%
ell <- function(x, y, beta) {
   n = length(x)
   loss = t(y - x%*%beta)%*%(y - x%*%beta)/n
   return(loss)
}

# Shouldnt you have to redefine n????
ell.grad <- function(x, y, beta) {
# trying to T beta
  grad = 2*(t(x)%*%(x%*%beta - y))/n
  return(grad)
}

ell.opt.mb <- function(x, y, beta.ini = c(0, 0), mbsize = 50, setpsize = 0.05, maxsteps = 200, tol = 1e-4) {
  beta.opt = beta.ini
  beta.opt.old = beta.opt
  beta.path = beta.ini
  ell.path = ell(x, y, beta.ini)
  iter = 0
  update = 1e8
  
  while((iter < maxsteps) && (update >  tol)) {
    mb.ind = sample(1:n, mbsize)
    x.mb = as.matrix(x[mb.ind, ], length(mb.ind), 2)
    y.mb = as.matrix(y[mb.ind, ], length(mb.ind), 1)
    grad = ell.grad(x.mb, y.mb, beta.opt)
    step = setpsize*grad
    beta.opt = beta.opt - step
    update = abs(beta.opt - beta.opt.old)/max(1e-4, abs(beta.opt.old))
    beta.opt.old = beta.opt
    beta.path = cbind(beta.path, beta.opt)
    ell.path = c(ell.path, ell(x, y, beta.opt))
    iter = iter + 1
  }
  
  obj = list(beta.opt = beta.opt, beta.path = beta.path, ell.path = ell.path)
  return(obj)
}
# x[c(2, 3, 4),]
# x.mb = as.matrix(x[mb.ind, ], length(1), 1)
```
